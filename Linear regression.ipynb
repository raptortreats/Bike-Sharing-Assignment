## Data Transformation

Lets use one hot coding method on the columns month, season, weathersit and weekday because they are categorical nominal variables

```python
df['season']=df['season'].astype('category')
df['weathersit']=df['weathersit'].astype('category')
df['mnth']=df['mnth'].astype('category')
df['weekday']=df['weekday'].astype('category')
```

```python
df
```

```python
dummies = pd.get_dummies(df[['mnth','season', 'weekday', 'weathersit']],drop_first=True)
```

```python
dummies
```

```python
df = pd.concat([df, dummies], axis=1)
```

```python
df=df.drop(columns=['mnth','season', 'weekday', 'weathersit'])
```

## split the data

importing the required libraries for model building

```python
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import RFE
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
train, test = train_test_split(df, train_size=0.7, random_state=100)
```

```python
train.shape
```

```python
test.shape
```

```python
train
```

### Rescaling the features

```python
scaler = MinMaxScaler()
```

```python
num_var = ['temp', 'atemp', 'hum', 'windspeed', 'cnt']
```

```python
train[num_var] = scaler.fit_transform(train[num_var])
train.head(10)
```

```python
y_train = train.pop('cnt')
```

```python
X_train = train
```

## RFE (Recursive Feature Elimination)

```python
lm = LinearRegression()
lm.fit(X_train, y_train)
rfe = RFE(lm, n_features_to_select=15)
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
col = X_train.columns[rfe.support_]
col
```

```python
X_train_rfe = X_train[col]
```

```python
vif = pd.DataFrame()

vif['Features'] = X_train_rfe.columns
vif['VIF'] = [variance_inflation_factor(X_train_rfe.values, i) for i in range(X_train_rfe.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
X_train_sm = sm.add_constant(X_train_rfe)
lr = sm.OLS(y_train, X_train_sm)
```

```python
lr_model = lr.fit()
lr_model.params
```

```python
lr_model.summary()
```

```python
vif = pd.DataFrame()

vif['Features'] = X_train_rfe.columns
vif['VIF'] = [variance_inflation_factor(X_train_rfe.values, i) for i in range(X_train_rfe.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```
